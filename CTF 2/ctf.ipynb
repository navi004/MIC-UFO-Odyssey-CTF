## African Cigarette Prices

```python
import pandas as pd

file2 = r"C:\Users\nvkaj\Pictures\Learn\MIC\CTF 2\Africa_Cigarette_Prices\cigarrette.csv"

# Try reading with different encodings
try:
    df2 = pd.read_csv(file2, encoding='utf-8')
except UnicodeDecodeError:
    df2 = pd.read_csv(file2, encoding='latin1')

# Display the dataframe
df2
```

```python
df2.columns
```

### 1.If you were tracking cigarette expenditures in Namibia, how much money, in local currency, would have been spent on cigarettes in total?

```python
namibia_df = df2[df2['Country'] == 'Namibia']
namibia_df
```

```python
round((namibia_df['Quantity'] * namibia_df['Local_Price_Per_Stick_Cigarette']).sum())
```

### 2.Lets say ,you are the finance minister of Lesotho and you have to find the total amount of money spent by your citizens on 18th jan 2016 for a specific reason on cigarettes , what would be your answer?

```python
ld_df = df2[(df2['Country'] == 'Lesotho') & (df2['Day'] == 18) & (df2['Month'] == 1) & (df2['Year'] == 2016)] 
ld_df
```

```python
(ld_df['Quantity']*ld_df['Local_Price']).sum()
```

### 3.You are a chain smoker in the suburb of Windhoek West and you are on a journey to find the total no of brand of cigarettes you can smoke that can be bought in a retail store?

```python
brands = df2[(df2['Suburb'] == 'Windhoek West') & (df2['Outlet_Type'] == 'Retail')]['Brand'].unique()
len(brands)
```
