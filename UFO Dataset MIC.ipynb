### Q7) This is green™©.

### green™© touched down on Earth around 9 PM on April 7, 2010. He zipped around the eastern US, keeping to the east of Colorado in his circular ride. By day's end, he zoomed back to his starting point. How far did green™© travel in total?

```python
d['month'] = d['datetime'].dt.month
d['day'] = d['datetime'].dt.day
```

```python
colorado_lon = -102
df7 = d[(d['month'] == 4) & (d['year'] == 2010) & (d['day'] == 7)]
df8 = df7[(df7['country'] == 'us') & (df7['shape'] == 'circle')]
df9 = df8[df8['longitude']>colorado_lon]
df10 = df9[df9['datetime'].dt.hour >= 21]
df10
```

```python
path = dict()
path['latitude'] = df10['latitude']
path['longitude'] = df10['longitude']
```

```python
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the Haversine distance between two points
    """
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))

    radius_earth_km = 6371

    distance = radius_earth_km * c
    return distance

latitudes = path['latitude'].values
longitudes = path['longitude'].values

total_distance = 0

for i in range(len(latitudes) - 1):
    distance = haversine(latitudes[i], longitudes[i], latitudes[i+1], longitudes[i+1])
    total_distance += distance

print(total_distance)
```
